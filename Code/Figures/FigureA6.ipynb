```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

plt.rcParams["font.family"] = "sans-serif"
```

```python
cd = "C://Users//scottjmk//Dropbox//Research//petra_moser//"

df1 = pd.read_csv(cd+"Replication//Data//MoS1956/MoS1956_Final.csv")
df1 = df1[["unique_id", "firstname", "lastname", "birthyear", "yearofmarriage", "child", "gender_clean", "vol_num", "see"]]

df2 = df1[df1["see"].isnull()][~df1["birthyear"].isnull()]
    
married_men = df2[df2["gender_clean"] == "male"][~df2["yearofmarriage"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "married_men"})
unmarried_men = df2[df2["gender_clean"] == "male"][df2["yearofmarriage"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "unmarried_men"})
married_women = df2[df2["gender_clean"] == "female"][~df2["yearofmarriage"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "married_women"})
unmarried_women = df2[df2["gender_clean"] == "female"][df2["yearofmarriage"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "unmarried_women"})

df = married_men.copy()
for x in [unmarried_men, married_women, unmarried_women]:
    df = pd.merge(df, x, on = "birthyear", how = "outer")

df.replace(np.nan, 0, inplace = True)
df = df.sort_values(by = "birthyear")

df["married_men_share"] = 100*df["married_men"]/(df["married_men"]+df["unmarried_men"])
df["married_women_share"] = 100*df["married_women"]/(df["married_women"]+df["unmarried_women"])
df.replace(np.nan, 0, inplace = True)

df
```

```python
fig, ax1 = plt.subplots(figsize = (5.5, 4))

df.set_index("birthyear")["married_women_share"].plot(ax = ax1, color = "black", linewidth = 1.5, label = "Women")
df.set_index("birthyear")["married_men_share"].plot(ax = ax1, color = "black", linewidth = 1.5, label = "Men", linestyle = "--")

ax1.set_ylabel("Married/Scientists (in %)", fontsize = 12)
ax1.set_xlabel("", fontsize = 12)
ax1.set_xlim(1890, 1930)
ax1.set_ylim(0, 100)
ax1.legend(loc = "upper center", bbox_to_anchor = (0.50, -0.15), ncol = 2, frameon = False, fontsize = 12)

ax1.tick_params(axis='both', which='major', labelsize=12)

ax1.yaxis.set_major_formatter(mtick.PercentFormatter(decimals = 0.1))
ax1.set_xticks([1890, 1900, 1910, 1920, 1930])

plt.savefig(cd+"Replication//Output//Figures/FigureA6A.jpg", transparent=False, bbox_inches="tight")
```

```python
df1 = pd.read_csv(cd+"Replication//Data//MoS1956/MoS1956_Final.csv")
df1 = df1[["unique_id", "firstname", "lastname", "birthyear", "yearofmarriage", "child", "gender_clean", "vol_num", "see"]]

df2 = df1[df1["see"].isnull()][~df1["birthyear"].isnull()][~df1["yearofmarriage"].isnull()]

df2["ageatmarriage"] = df2["yearofmarriage"]-df2["birthyear"]
    
married_men = df2[df2["gender_clean"] == "male"].groupby("birthyear")["ageatmarriage"].mean().reset_index().rename(columns = {"ageatmarriage": "ageatmarriage_men"})
married_women = df2[df2["gender_clean"] == "female"].groupby("birthyear")["ageatmarriage"].mean().reset_index().rename(columns = {"ageatmarriage": "ageatmarriage_women"})

df = pd.merge(married_men, married_women, on = "birthyear", how = "outer")

df.replace(np.nan, 0, inplace = True)
df = df.sort_values(by = "birthyear")

df
```

```python
fig, ax1 = plt.subplots(figsize = (5.5, 4))

df.set_index("birthyear")["ageatmarriage_women"].plot(ax = ax1, color = "black", linewidth = 1.5, label = "Women")
df.set_index("birthyear")["ageatmarriage_men"].plot(ax = ax1, color = "black", linewidth = 1.5, label = "Men", linestyle = "--")

ax1.set_ylabel("Age at Marriage", fontsize = 12)
ax1.set_xlabel("", fontsize = 12)
ax1.set_xlim(1890, 1930)
ax1.set_xticks([1890, 1900, 1910, 1920, 1930])
ax1.set_ylim(20, 35)
ax1.legend(loc = "upper center", bbox_to_anchor = (0.50, -0.15), ncol = 2, frameon = False, fontsize = 12)

ax1.tick_params(axis='both', which='major', labelsize=12)

plt.savefig(cd+"Replication//Output//Figures/FigureA6B.jpg", transparent=False, bbox_inches="tight")
```

```python
df1 = pd.read_csv(cd+"Replication//Data//MoS1956/MoS1956_Final.csv")
df1 = df1[["unique_id", "firstname", "lastname", "birthyear", "child", "gender_clean", "vol_num", "see"]]

df2 = df1[df1["see"].isnull()][~df1["birthyear"].isnull()]
    
fathers = df2[df2["gender_clean"] == "male"][~df2["child"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "fathers"})
other_men = df2[df2["gender_clean"] == "male"][df2["child"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "other_men"})
mothers = df2[df2["gender_clean"] == "female"][~df2["child"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "mothers"})
other_women = df2[df2["gender_clean"] == "female"][df2["child"].isnull()].groupby("birthyear").size().reset_index().rename(columns = {0: "other_women"})

df = fathers.copy()
for x in [other_men, mothers, other_women]:
    df = pd.merge(df, x, on = "birthyear", how = "outer")

df.replace(np.nan, 0, inplace = True)
df = df.sort_values(by = "birthyear")

df["fathers_share"] = 100*df["fathers"]/(df["fathers"]+df["other_men"])
df["mothers_share"] = 100*df["mothers"]/(df["mothers"]+df["other_women"])

df.replace(np.nan, 0, inplace = True)

df
```

```python
fig, ax1 = plt.subplots(figsize = (5.5, 4))

df.set_index("birthyear")["mothers_share"].plot(ax = ax1, color = "black", linewidth = 1.5, label = "Women")
df.set_index("birthyear")["fathers_share"].plot(ax = ax1, color = "black", linewidth = 1.5, label = "Men", linestyle = "--")

ax1.set_ylabel("Parents/Scientists (in %)", fontsize = 12)
ax1.set_xlabel("", fontsize = 12)
ax1.set_xlim(1890, 1930)
ax1.set_ylim(0, 100)
ax1.legend(loc = "upper center", bbox_to_anchor = (0.50, -0.15), ncol = 2, frameon = False, fontsize = 12)

ax1.tick_params(axis='both', which='major', labelsize=12)
ax1.set_xticks([1890, 1900, 1910, 1920, 1930])
ax1.yaxis.set_major_formatter(mtick.PercentFormatter(decimals = 0.1))

plt.savefig(cd+"Replication//Output//Figures/FigureA6C.jpg", transparent=False, bbox_inches="tight")
```
