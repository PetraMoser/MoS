```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

plt.rcParams["font.family"] = "sans-serif"
```

```python
cd = "C://Users//scottjmk//Dropbox//Research//petra_moser//"

df1 = pd.read_csv(cd+"Replication//Data//PanelData/PanelData_Final.csv")

df2 = df1[df1["age"]>=18][df1["age"]<=65][df1["vol_num"]==1]

df2["patent"] = 1000*df2["patent"]

mothers = df2[df2["female_parent"]==1]
other_women = df2[df2["female_nonparent"]==1]
fathers = df2[df2["male_parent"]==1]
other_men = df2[df2["male_nonparent"]==1]

mothers_count = mothers.groupby("age").size().reset_index().rename(columns={0:"mothers_count"})
other_women_count = other_women.groupby("age").size().reset_index().rename(columns={0:"other_women_count"})
fathers_count = fathers.groupby("age").size().reset_index().rename(columns={0:"fathers_count"})
other_men_count = other_men.groupby("age").size().reset_index().rename(columns={0:"other_men_count"})

mothers_patent = mothers[["age", "patent"]].groupby("age").sum().reset_index().rename(columns={"patent":"mothers_total"})
other_women_patent = other_women[["age", "patent"]].groupby("age").sum().reset_index().rename(columns={"patent":"other_women_total"})
fathers_patent = fathers[["age", "patent"]].groupby("age").sum().reset_index().rename(columns={"patent":"fathers_total"})
other_men_patent = other_men[["age", "patent"]].groupby("age").sum().reset_index().rename(columns={"patent":"other_men_total"})

df = pd.merge(mothers_count, mothers_patent, on = "age")
df = pd.merge(df, other_women_count, on = "age")
df = pd.merge(df, other_women_patent, on = "age")
df = pd.merge(df, fathers_count, on = "age")
df = pd.merge(df, fathers_patent, on = "age")
df = pd.merge(df, other_men_count, on = "age")
df = pd.merge(df, other_men_patent, on = "age")

df["mothers_avg"] = df["mothers_total"]/df["mothers_count"]
df["fathers_avg"] = df["fathers_total"]/df["fathers_count"]
df["other_women_avg"] = df["other_women_total"]/df["other_women_count"]
df["other_men_avg"] = df["other_men_total"]/df["other_men_count"]

columns = ["mothers_avg", "fathers_avg", "other_women_avg", "other_men_avg"]

for x in columns:
    df[x] = df[x].rolling(center=True, window=3).mean()

df
```

```python
fig, ax1 = plt.subplots(figsize = (5.5, 4))
ax2 = ax1.twinx()

df.set_index("age")["mothers_avg"].plot(ax = ax1, color = "black", linewidth = 1.5, label = "Mothers")
df.set_index("age")["fathers_avg"].plot(ax = ax2, color = "black", linewidth = 1.5, linestyle = "--", label = "Fathers")

ax1.set_xlim(18, 65)
ax2.set_ylim(0, 200)
ax1.set_ylim(0, 80)
ax1.set_yticks([0, 20, 40, 60, 80])
ax2.set_yticks([0, 50, 100, 150, 200])
ax1.set_xlabel("", fontsize = 10)
ax2.set_ylabel("Patents by Fathers", fontsize = 10)
ax1.set_ylabel("Patents by Mothers", fontsize = 10)

ax1.legend(loc = "upper center", bbox_to_anchor = (0.35, -0.125), frameon = False, fontsize = 10)
ax2.legend(loc = "upper center", bbox_to_anchor = (0.65, -0.125), frameon = False, fontsize = 10)

ax1.tick_params(axis='both', which='major', labelsize=10)
ax2.tick_params(axis='both', which='major', labelsize=10)
ax1.axvline(x = 27, linewidth = 1.25, color = "red")

plt.savefig(cd+"Replication//Output//Figures/Figure1A.jpg", transparent=False, bbox_inches="tight")
```
